```python
import pandas as pd
import numpy as np

# Load the Excel file
file_path = '/content/drive/MyDrive/Modelin and Sensitivity_Bass Connections/DHPP_data.xlsx'
data = pd.read_excel(file_path)
#data = data.drop(columns=['Unnamed: 14','Unnamed: 15', 'Unnamed: 16','Unnamed: 17'])
data
```

```python
# Rename columns for consistency (update these if needed)
data.rename(columns={'Major Area': 'Prefecture',
                     'Status': 'Status',
                     'DHPP': 'DHPP'}, inplace=True)
```

```python
# **Fix: Group and Aggregate by Status and Prefecture**
dhpp_summary = data.groupby(['Prefecture', 'Status'])['DHPP'].sum().reset_index()
dhpp_summary
```

```python
# Create separate grouped DHPP values for each year
# 2025: Only "Operational"
dhpp_2025 = dhpp_summary[dhpp_summary['Status'] == 'Operational']\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2025.rename(columns={'DHPP': 'DHPP_2025'}, inplace=True)
dhpp_2025
```

```python
# 2030: "Operational" + "Under Construction"
dhpp_2030 = dhpp_summary[dhpp_summary['Status'].isin(['Operational', 'Under Construction'])]\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2030.rename(columns={'DHPP': 'DHPP_2030'}, inplace=True)
dhpp_2030
```

```python
# 2050: "Operational" + "Under Construction" + "Announced" + "Preconstruction"
dhpp_2050 = dhpp_summary[dhpp_summary['Status'].isin(['Operational', 'Under Construction', 'announced', 'pre-construction'])]\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2050.rename(columns={'DHPP': 'DHPP_2050'}, inplace=True)
dhpp_2050
```

```python
# Merge results to ensure different values are correctly assigned
dhpp_final_summary = dhpp_2025\
    .merge(dhpp_2030, on='Prefecture', how='outer')\
    .merge(dhpp_2050, on='Prefecture', how='outer')

# Replace NaN values with 0
dhpp_final_summary.fillna(0, inplace=True)

dhpp_final_summary.sort_values('DHPP_2050')
```

```python
# Save the final summary to a file for ArcGIS
output_file = "DHPP_Summary_by_Prefecture.xlsx"
dhpp_final_summary.to_excel(output_file, index=False)
```
