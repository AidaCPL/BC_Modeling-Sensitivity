```python
df_iaea = pd.read_excel('/content/drive/MyDrive/Modelin and Sensitivity_Bass Connections/DHPP calculation with HDD data/IAEA dataset - no HD.xlsx')

df_iaea
```

```python
df_iaea['load_factor'] = df_iaea['Average Lifetime Load Factor (%)']/100
```

```python
df_iaea = df_iaea.drop(columns= 'Average Lifetime Load Factor (%)')
```

```python
hdd_data = pd.read_excel('/content/drive/MyDrive/Modelin and Sensitivity_Bass Connections/DHPP calculation with HDD data/City HDD until 2024-06-30.xlsx')

hdd_data
```

```python
hdd_data.columns = hdd_data.iloc[0]  # Set the first row as column names
hdd_data = hdd_data[1:].reset_index(drop=True)
```

```python
hdd_data
```

```python
df_iaea['Prefecture'].unique()
```

```python
# List of columns to keep
prefectures = ['Fuzhou', 'Ningde', 'Zhangzhou', 'Huizhou', 'Jiangmen', 'Maoming',
       'Shanwei', 'Shenzhen', 'Yangjiang', 'Zhanjiang', 'Fangchenggang',
       'Changjiang', 'Cangzhou', 'Lianyungang', 'Dalian', 'Huludao',
       'Weihai', 'Yantai', 'Jiaxing', 'Ningbo', 'Taizhou', 'Wenzhou']

# Filter the DataFrame to keep only the specified columns
hdd_data = hdd_data[prefectures]
```

```python
import pandas as pd

# Convert the wide format DataFrame to a long format using melt
hdd_long = hdd_data.melt(var_name='Prefecture', value_name='HDD 2012-2014')

# Compute the mean HDD for each prefecture
hdd_data_4527days = hdd_long.groupby('Prefecture', as_index=False)['HDD 2012-2014'].sum()

hdd_data_4527days
```

```python
# Merge using a left join to keep all plants and repeat HDD values
df_merged = df_iaea.merge(hdd_data_4527days, on="Prefecture", how="left")
df_merged
```

```python
df_merged['DHPP [MW_th]'] = df_merged['Reference Unit Power [MW]']* df_merged['load_factor']* (df_merged['HDD 2012-2014']/4527) * 1.6
```

```python
df_merged
```

# Estimating Production in GWh a year

$DHPP [MW_th]$ --> a day (average HDD a day)

$DHPP [MW_th] \times 8.76$ (hours in a year/ 1000)

```python
df_merged['Estimated yearly energy production (GWh)'] = df_merged['DHPP [MW_th]']* 8.76
```

```python
df_merged
```

```python
# Save the final summary to a file for ArcGIS
output_file = "DHPP_HDD_data.xlsx"
df_merged.to_excel(output_file, index=False)
```

## Grouping plants per prefecture for visualization

```python
#Group and Aggregate by Status and Prefecture
dhpp_summary = df_merged.groupby(['Prefecture', 'Status'])['DHPP'].sum().reset_index()
dhpp_summary
```

```python
# Create separate grouped DHPP values for each year
# 2025: Only "Operational"
dhpp_2025 = dhpp_summary[dhpp_summary['Status'] == 'Operational']\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2025.rename(columns={'DHPP': 'DHPP_2025'}, inplace=True)
dhpp_2025
```

```python
# 2030: "Operational" + "Under Construction"
dhpp_2030 = dhpp_summary[dhpp_summary['Status'].isin(['Operational', 'Under Construction'])]\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2030.rename(columns={'DHPP': 'DHPP_2030'}, inplace=True)
dhpp_2030
```

```python
# 2050: "Operational" + "Under Construction" + "Announced" + "Preconstruction"
dhpp_2050 = dhpp_summary[dhpp_summary['Status'].isin(['Operational', 'Under Construction', 'announced', 'pre-construction'])]\
    .groupby('Prefecture')['DHPP'].sum().reset_index()
dhpp_2050.rename(columns={'DHPP': 'DHPP_2050'}, inplace=True)
dhpp_2050
```

```python
# Merge results to ensure different values are correctly assigned
dhpp_final_summary = dhpp_2025\
    .merge(dhpp_2030, on='Prefecture', how='outer')\
    .merge(dhpp_2050, on='Prefecture', how='outer')

# Replace NaN values with 0
dhpp_final_summary.fillna(0, inplace=True)

dhpp_final_summary.sort_values('DHPP_2025')
```
